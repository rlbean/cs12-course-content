## Part 1 — Common Escape Characters

```python
# \n  — newline
print("Line 1\nLine 2\nLine 3")
```

```python
# \t  — tab
print("Name:\tAlice")
print("Score:\t95")
print("Grade:\tA")
```

```python
# \\  — prints a real backslash
print("C:\\Users\\rbean\\Documents\\test.txt") 
# You could replace my name for yours here and have the system create this file. 
```

```python
# \"  and  \'  — quotes inside strings
print("She said \"Hello, World!\"")
print('It\'s a great day to code.')
```

```python
# Combining escape characters
print("Item\t\tPrice")
print("Apple\t\t$1.00")
print("Banana\t\t$0.50")
print("Cherry\t\t$2.75")
```

## Part 2 — Unicode with `\u`

```python
# Basic \uXXXX examples
print("\u2605")   # ★ star
print("\u00A9")   # © copyright
print("\u03C0")   # π pi
print("\u221E")   # ∞ infinity
print("\u2713")   # ✓ checkmark
print("\u2717")   # ✗ cross
```

```python
# Unicode characters inside strings
print("Math: \u03C0 \u2248 3.14159")
print("Temperature: 22 \u00B0C")
print("Result: 5 \u2260 6")
print("Range: 1 \u2264 x \u2264 10")
```

```python
# \U (capital U) — 8 digits — used for emoji
print("\U0001F600")   # 😀
print("\U0001F40D")   # 🐍
print("\U0001F3AE")   # 🎮
print("\U0001F4BB")   # 💻
```

## Part 3 — Finding Unicode Codes

If you already have a character, Python can tell you its code.

```python
# ord() gives the decimal number for a character
# hex() converts that to the hex code used in \u

char = "★"
print("Character :", char)
print("Decimal   :", ord(char))
print("Hex       :", hex(ord(char)))   # 0x2605 → \u2605
```

```python
# Try it with a few characters
characters = ["A", "a", "©", "π", "✓", "∞"]

for c in characters:
    print(f"{c}  →  decimal: {ord(c):>6}   hex: {hex(ord(c))}")
```

## Part 4 — Box-Drawing Characters

```python
# Print each box-drawing character with its code
box_chars = [
    ("\u2554", "top-left corner"),
    ("\u2557", "top-right corner"),
    ("\u255A", "bottom-left corner"),
    ("\u255D", "bottom-right corner"),
    ("\u2550", "horizontal line"),
    ("\u2551", "vertical line"),
    ("\u2560", "left T-junction"),
    ("\u2563", "right T-junction"),
]

for char, name in box_chars:
    print(f"  {char}   {name}")
```

```python
# Build a simple box
top    = "\u2554" + "\u2550" * 22 + "\u2557"
middle = "\u2551" + "  Welcome to Python!  " + "\u2551"
bottom = "\u255A" + "\u2550" * 22 + "\u255D"

print(top)
print(middle)
print(bottom)
```

```python
# A box with a divider line
H = "\u2550"   # horizontal
V = "\u2551"   # vertical
TL = "\u2554"  # top-left
TR = "\u2557"  # top-right
BL = "\u255A"  # bottom-left
BR = "\u255D"  # bottom-right
LT = "\u2560"  # left T
RT = "\u2563"  # right T

width = 24

print(TL + H * width + TR)
print(V  + "  STUDENT REPORT CARD   " + V)
print(LT + H * width + RT)
print(V  + "  Math:    95           " + V)
print(V  + "  English: 88           " + V)
print(V  + "  Science: 91           " + V)
print(BL + H * width + BR)
```

## Part 5 — Putting It All Together

```python
# A styled text menu using escape characters + unicode

star  = "\u2605"
arrow = "\u2192"
check = "\u2713"
H  = "\u2550"
V  = "\u2551"
TL = "\u2554"
TR = "\u2557"
BL = "\u255A"
BR = "\u255D"
LT = "\u2560"
RT = "\u2563"

w = 28

print(TL + H * w + TR)
print(V  + f"   {star} ADVENTURE GAME {star}      " + V)
print(LT + H * w + RT)
print(V  + f"   {arrow} 1. New Game            " + V)
print(V  + f"   {arrow} 2. Load Save           " + V)
print(V  + f"   {arrow} 3. Settings            " + V)
print(V  + f"   {check} 4. Quit                " + V)
print(BL + H * w + BR)
```
