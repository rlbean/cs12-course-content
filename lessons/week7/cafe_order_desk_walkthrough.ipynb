## Part 1 - The Big Rule: Return, Don't Print

In C#, you might have printed inside a method. In this assignment, **that's not allowed.**

Functions do the **work**. `main()` does the **talking**.

| C# habit | Python rule for this assignment |
|---|---|
| `Console.WriteLine()` inside a method | Not allowed inside helper functions |
| `return value;` from a method | This is the only way out |
| Call method → use result in `Main()` | `main()` gets the return value and prints it |

Think of it like a **vending machine**:  
You press a button → the machine does its work → it **gives you the item back**.  
The machine doesn't shout at you - *you* look at what came out.

## Part 2 - A Bad Example (Don't Do This)

Here's what **NOT** to do. The function prints its own result:

````python
# BAD - the function prints its own result
def get_drink_price_bad(choice):
    if choice == 1:
        print("Price is $3.50")   # ← WRONG. Function is doing the talking.
    elif choice == 2:
        print("Price is $4.00")

get_drink_price_bad(1)
````

Why is this a problem?
- You can **never reuse** this function for a calculation - it only talks, it doesn't give anything back
- You can't store the result in a variable
- You can't pass the result to another function

---

## Part 3 - The Correct Way: Return a Value

````python
# GOOD - the function returns its result
def get_drink_price(choice):
    """Takes a menu choice (int). Returns the drink price (float)."""
    if choice == 1:
        return 3.50      # ← Sends the value BACK to whoever called this function
    elif choice == 2:
        return 4.00
    elif choice == 3:
        return 5.25
    else:
        return 0.0       # ← Always have a fallback return

# Now main() is in charge of printing:
price = get_drink_price(2)          # Store the returned value
print(f"The price is ${price:.2f}") # main() (or whoever called it) prints
````

Notice:
- `get_drink_price()` has **no print statement** inside it
- The caller stores the result in `price`
- The caller decides what to do with it (print it, pass it somewhere else, etc.)

## Part 4 - Passing a Return Value into Another Function

This is where it gets powerful. One function's **output becomes another function's input**.

````python
def calculate_total(price, quantity):
    """Takes a price (float) and quantity (int). Returns the total cost (float)."""
    total = price * quantity
    return total          # ← Return the result, don't print it!

# --- This is how main() chains functions together ---

# Step 1: get the price from one function
drink_price = get_drink_price(1)          # returns 3.50

# Step 2: pass that price into another function
order_total = calculate_total(drink_price, 3)  # 3.50 × 3 = 10.50

# Step 3: main() prints the result
print(f"3 drinks × ${drink_price:.2f} = ${order_total:.2f}")
````

## Part 5 - The `while True` / `break` Menu Loop

This is the **engine** of Assignment 4. The menu keeps showing until the user chooses to quit.

In C#:  
```csharp
while (true) {
    // show menu, get choice
    if (choice == 4) break;
}
```

In Python, it's almost identical:

````python
def show_menu():
    """Returns the menu as a formatted string. Does NOT print it."""
    menu = (
        "\n=== Cafe Order Desk ==="
        "\n1. Check drink price"
        "\n2. Calculate order total"
        "\n3. Check for discount"
        "\n4. Exit"
        "\nEnter your choice: "
    )
    return menu    # Return the string - main() will print it

# Preview what the menu looks like
print(show_menu())
````

Now here's the loop. Notice **every** `print()` call is inside `main()`:

````python
# Run this cell to try the full Cafe Order Desk!
import random

# ─────────────────────────────────────────────
# HELPER FUNCTIONS  (no print statements here!)
# ─────────────────────────────────────────────

def show_menu():
    """Returns the formatted menu string."""
    return (
        "\n=== Cafe Order Desk ==="
        "\n1. Check drink price"
        "\n2. Calculate order total"
        "\n3. Check for a discount"
        "\n4. Exit"
        "\nYour choice: "
    )

def get_drink_price(choice):
    """Takes a drink choice (int 1-3). Returns the price (float)."""
    if choice == 1:
        return 3.50    # Espresso
    elif choice == 2:
        return 4.75    # Latte
    elif choice == 3:
        return 5.25    # Specialty drink
    else:
        return 0.0     # Invalid choice fallback

def calculate_total(price, quantity):
    """Takes price (float) and quantity (int). Returns total cost (float)."""
    return price * quantity

def apply_discount(total, quantity):
    """If quantity >= 3, applies a 10% discount. Returns the discounted total (float)."""
    if quantity >= 3:
        return total * 0.90    # 10% off
    else:
        return total           # No change — still return something!

def get_daily_special():
    """Picks a random daily special. Returns the special name (str)."""
    specials = ["Caramel Macchiato", "Iced Matcha", "Honey Oat Latte", "Cold Brew Float"]
    index = random.randint(0, len(specials) - 1)
    return specials[index]

# ─────────────────────────────────────────────
# MAIN — only place allowed to print!
# ─────────────────────────────────────────────

def main():
    print("Welcome to the Cafe Order Desk!")
    print(f"Today's special: {get_daily_special()}")   # call function, print result

    while True:                                         # keep looping until break
        choice = input(show_menu())                     # show_menu() returns the string

        if choice == "1":
            print("\nDrinks: 1-Espresso  2-Latte  3-Specialty")
            drink = int(input("Pick a drink (1-3): "))
            price = get_drink_price(drink)              # ← store return value
            print(f"Price: ${price:.2f}")               # ← main() prints it

        elif choice == "2":
            price = float(input("Enter the drink price: $"))
            qty   = int(input("How many drinks? "))
            total = calculate_total(price, qty)         # ← store return value
            print(f"Total before discount: ${total:.2f}")

        elif choice == "3":
            price = float(input("Enter drink price: $"))
            qty   = int(input("How many drinks? "))
            total = calculate_total(price, qty)         # chain: total feeds into discount
            final = apply_discount(total, qty)          # ← store return value
            if qty >= 3:
                print(f"Discount applied! ${total:.2f} → ${final:.2f}")
            else:
                print(f"No discount yet. Total: ${final:.2f}")

        elif choice == "4":
            print("Thanks for visiting the Cafe! Goodbye.")
            break                                       # ← exits the while True loop

        else:
            print("Invalid choice. Please enter 1-4.")

main()
````

## Part 6 - Let's Trace the Flow

When the user picks **option 3**, here's what happens step by step:

```
main()
  │
  ├─ input()          ← user types "3"
  │
  ├─ calculate_total(price, qty)
  │       │
  │       └─ returns total  →  stored in variable 'total'
  │
  ├─ apply_discount(total, qty)
  │       │
  │       └─ returns final  →  stored in variable 'final'
  │
  └─ print(...)       ← main() prints the result
```

**Every function gives something back. main() is the one doing the talking.**

## Part 7 - Common Mistakes to Avoid

| Mistake | Why it's wrong | Fix |
|---|---|---|
| `print()` inside a helper function | Can't reuse the result elsewhere | Use `return` instead |
| Forgetting `return` on one branch | Function returns `None` silently | Every `if/elif/else` needs a `return` |
| Not storing the return value | `get_drink_price(2)` on its own - value is lost | `price = get_drink_price(2)` |
| Using `exit()` instead of `break` | Crashes the whole program | Use `break` to leave the `while True` loop |
| Mixing up parameter names | Confusing variable scope | Keep parameter names clear and local |

## Summary - The Three Rules

1. **Functions return, they don't print.**

2. **`main()` is the only place that prints.**
