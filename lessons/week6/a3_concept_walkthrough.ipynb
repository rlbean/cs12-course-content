## 2 - Functions That Return Values

In Assignment 3, your program needs **functions** - small reusable blocks of code.

The important rule:
> **Functions should *return* a result. `main()` handles all the printing.**

```python
def my_function():
    # do some work...
    return result           # sends the answer back to whoever called this

def main():
    answer = my_function()  # store the returned value
    print(answer)           # main() does the printing
```

Think of a function like a vending machine:
- You press a button (call the function)
- It does its job quietly inside
- It gives something back to you (the `return` value)

Our `spin_wheel()` function picks a random colour and **returns** it - it doesn't print anything itself.

````python
import random

# This function picks a random colour and returns it.
# Notice: NO print() inside this function.
def spin_wheel():
    section = random.randint(1, 4)
    if section == 1:
        return "Red"
    elif section == 2:
        return "Blue"
    elif section == 3:
        return "Green"
    else:
        return "Yellow"

# main() calls the function and handles the output
def main():
    colour = spin_wheel()             # call it, store the return value
    print("You landed on:", colour)   # main() does the printing

main()
````

## 3 - Validating User Input

Assignment 3 requires checking that the user enters a valid number.  
The standard pattern is a `while True` loop - it keeps asking until it gets a good answer.

### The Pattern
```python
while True:
    user_input = input("Enter something: ")

    if user_input.isdigit():     # is it a number at all?
        num = int(user_input)    # convert to integer
        if 1 <= num <= 10:       # is it in range?
            return num           # valid - return it, loop ends

    # if we reached here, input was bad
    print("Invalid - try again.")
```

### `.isdigit()` - what does it do?
Checks if every character in the string is a digit (0–9). Returns `True` or `False`.
Protects against crashes when the user types letters, symbols, or negative numbers.

| Input | `.isdigit()` |
|---|---|
| `"5"` | `True` |
| `"abc"` | `False` |
| `"-3"` | `False` |
| `"3.5"` | `False` |

````python
# isdigit() demo
print("5".isdigit())     # True
print("abc".isdigit())   # False
print("-3".isdigit())    # False
print("10".isdigit())    # True
````

````python
# Full input validation: ask how many spins (1 to 8)
def get_num_spins():
    while True:
        user_input = input("How many spins would you like? (1-8): ")

        if user_input.isdigit():           # step 1: is it a number?
            num = int(user_input)          # step 2: convert it
            if 1 <= num <= 8:              # step 3: is it in range?
                return num                 # valid!

        print("  Please enter a whole number between 1 and 8.")

def main():
    spins = get_num_spins()
    print(f"Great - spinning {spins} time(s).")

main()
# Try entering: letters, a negative number, 0, 9, then a valid number
````

## 4 - Accumulating a Score with a For Loop

Once we know how many rounds to run, we use a `for` loop to repeat that many times and build up a total.

```python
for i in range(1, num_spins + 1):
    # i = 1, then 2, then 3... up to num_spins
```

### Why `range(1, num_spins + 1)`?
- `range(start, stop)` goes from `start` up to, but **NOT** including, `stop`
- Starting at `1` lets us use `i` as the round number in output
- `+1` ensures we get exactly `num_spins` iterations

### The accumulator pattern
```python
# start at 0 - BEFORE the loop
total_score = 0                 

for i in range(1, num_spins + 1):
    colour = spin_wheel()
    points = get_points(colour)
    # add this round to the running total
    total_score += points       

# return the final total - AFTER the loop
return total_score              
```

````python
import random

def spin_wheel():
    section = random.randint(1, 4)
    if section == 1:
        return "Red"
    elif section == 2:
        return "Blue"
    elif section == 3:
        return "Green"
    else:
        return "Yellow"

def play_game(num_spins):
    # accumulator - starts at 0
    total_score = 0                         

    # repeat num_spins times
    for i in range(1, num_spins + 1):
        # get a result each round       
        colour = spin_wheel()               
        print(f"  Spin {i}: {colour}")
        # every colour is worth 10 (for now)
        total_score += 10                   

    # send total back to main()
    return total_score                      

def main():
    score = play_game(4)
    print(f"\nTotal Score: {score}")

main()
````

## 5 - Scoring with `if / elif / else`

Most Assignment 3 themes need different outcomes to award different points.  
The pattern is a separate function that receives a result and **returns** the point value:

```python
def get_points(colour):
    # rarest - check first, highest points
    if colour == "Red":      
        return 50
    elif colour == "Blue":
        return 30
    elif colour == "Green":
        return 20
    # everything else - lowest points
    else:                    
        return 10
```

**Order matters** - Python checks top to bottom and stops at the first match.  
Put the rarest / most specific condition first. The general `else` catches everything else.

You can also check two things at once using `and`:
```python
# both red - jackpot!
if colour_a == "Red" and colour_b == "Red":   
    return 100
```

````python
import random

def spin_wheel():
    section = random.randint(1, 4)
    if section == 1:
        return "Red"
    elif section == 2:
        return "Blue"
    elif section == 3:
        return "Green"
    else:
        return "Yellow"

# A scoring function - receives a colour, returns points
def get_points(colour):
    # rarest - check first!
    if colour == "Red":       
        return 50
    elif colour == "Blue":
        return 30
    elif colour == "Green":
        return 20
    # Yellow - most common, fewest points
    else:                     
        return 10

def play_game(num_spins):
    total_score = 0
    for i in range(1, num_spins + 1):
        colour = spin_wheel()
        # pass colour in, get points back
        points = get_points(colour)      
        total_score += points
        print(f"  Spin {i}: {colour} -> {points} pts")
    return total_score

def main():
    score = play_game(5)
    print(f"\nFinal Score: {score}")

    # Final message based on score
    if score >= 150:
        print("Amazing run!")
    elif score >= 80:
        print("Pretty good!")
    else:
        print("Better luck next time.")

main()
````

## 6 - The Full Program

Here is how all the pieces connect:

```
main()
  |
  +--- get_num_spins()      validates input, returns an integer
  |
  +--- play_game(n)         runs the loop, returns total score
  |         |
  |         +-- spin_wheel()    returns a random colour
  |         +-- get_points()    returns point value for that colour
  |
  +--- prints final results and outcome message
```

**Rules to remember:**
- `spin_wheel()` - returns a value, no printing
- `get_points()` - returns a value, no printing
- `get_num_spins()` - returns a value, only prints error messages
- `play_game()` - returns total score; printing each round result here is fine
- `main()` - calls everything, stores all return values, handles all final output

````python
'''
Name:        Bean (Demo)
Date:        March 2026
Assignment:  Colour Spinner - Teaching Demo
Description: A spinner game demonstrating random numbers, functions
             with return values, loops, input validation, and scoring.
'''

import random


# Spins the wheel. Returns a colour string.
def spin_wheel():
    section = random.randint(1, 4)
    if section == 1:
        return "Red"
    elif section == 2:
        return "Blue"
    elif section == 3:
        return "Green"
    else:
        return "Yellow"


# Returns the point value for a given colour.
def get_points(colour):
    if colour == "Red":
        return 50
    elif colour == "Blue":
        return 30
    elif colour == "Green":
        return 20
    else:
        return 10


# Asks how many spins (1-8). Validates input. Returns integer.
def get_num_spins():
    while True:
        user_input = input("How many spins would you like? (1-8): ")
        if user_input.isdigit():
            num = int(user_input)
            if 1 <= num <= 8:
                return num
        print("  Please enter a whole number between 1 and 8.")


# Runs the game loop. Returns the total score.
def play_game(num_spins):
    total_score = 0

    for i in range(1, num_spins + 1):
        colour = spin_wheel()
        points = get_points(colour)
        total_score += points
        print(f"  Spin {i}: {colour} -> {points} pts")

    return total_score


# Controls the program. All final output lives here.
def main():
    print("================================")
    print("        COLOUR SPINNER          ")
    print("================================\n")

    num_spins   = get_num_spins()
    print()

    total_score = play_game(num_spins)

    print(f"\n--- Final Score: {total_score} ---")

    if total_score >= 150:
        print("Amazing run - you're on fire!")
    elif total_score >= 80:
        print("Pretty good spin!")
    else:
        print("Better luck next time.")


main()
````
