## 1 - Generating Random Numbers

Python comes with a built-in `random` module. You have to **import** it before you can use it.

The function we care about most is:
```python
random.randint(low, high)   # returns a whole number between low and high (inclusive)
```

### C# vs Python comparison
| C# | Python |
|---|---|
| `Random rng = new Random();` | `import random` (at the top of the file) |
| `int n = rng.Next(1, 7);` | `n = random.randint(1, 6)` |

> ⚠️ **Watch out:** In C#, `Next(1, 7)` gives you 1–6 (upper bound is *exclusive*).  
> In Python, `randint(1, 6)` gives you 1–6 (upper bound is *inclusive*).  
> Same result - different rules!

````python
import random

# Roll a six-sided die
die = random.randint(1, 6)
print("Die roll:", die)

# Pick a random number between 1 and 100
big_number = random.randint(1, 100)
print("Big number:", big_number)

# Run this cell a few times - the numbers change every time!
````

## 2 - Functions That Return Values

In Assignment 3, your program needs **functions** - small reusable blocks of code.

The important rule for this assignment:
> **Functions should *return* a result. The `main()` function handles all the printing.**

Here is the pattern:

```python
def my_function():
    # do some work...
    return result       # <-- sends the answer back to whoever called this function

def main():
    answer = my_function()   # <-- stores the returned value
    print(answer)            # <-- main() does the printing
```

Think of a function like a vending machine:
- You press a button (call the function)
- It does its job
- It gives something back to you (the `return` value)

````python
import random

# --- This function generates a random coin flip result ---
# It does NOT print anything - it just returns the result
def flip_coin():
    number = random.randint(1, 2)   # 1 = Heads, 2 = Tails
    if number == 1:
        return "Heads"
    else:
        return "Tails"

# --- main() calls the function and prints the result ---
def main():
    result = flip_coin()        # call the function, store the return value
    print("Result:", result)    # main() handles the output

main()
````

## 3 - Validating User Input

Assignment 3 requires you to check that the user enters a valid number.  
The standard pattern for this is a `while True` loop with a `break`.

### The Pattern
```python
while True:
    user_input = input("Enter something: ")
    
    if user_input.isdigit():        # is it a number at all?
        num = int(user_input)       # convert it
        if 1 <= num <= 10:          # is it in range?
            return num              # ✅ valid - return it and exit the loop
    
    print("Invalid - try again.")   # ❌ if we get here, it was bad input
```

### Why `while True`?
- The loop runs *forever* until we say `return` (or `break`)
- That guarantees we keep asking until we get a good answer
- Clean, readable, and the standard Python approach

### `.isdigit()` - what does it do?
- Checks if every character in the string is a digit (0–9)
- Returns `True` or `False`
- Protects us from crashes when the user types letters or symbols

| Input | `.isdigit()` result |
|---|---|
| `"5"` | `True` |
| `"abc"` | `False` |
| `"-3"` | `False` (negative sign is not a digit) |
| `"3.5"` | `False` (decimal point is not a digit) |

````python
# --- isdigit() demo ---
print("5".isdigit())     # True
print("abc".isdigit())   # False
print("-3".isdigit())    # False
print("10".isdigit())    # True
````

````python
# --- Full input validation function ---
def get_num_flips():
    while True:
        user_input = input("How many flips would you like? (1–10): ")

        if user_input.isdigit():
            num = int(user_input)
            if 1 <= num <= 10:
                return num                        # ✅ valid input - exit loop

        print("  Please enter a whole number between 1 and 10.")

def main():
    flips = get_num_flips()
    print(f"Great - we will flip {flips} time(s).")

main()

# Try typing: letters, a negative number, 0, 11, and finally a valid number
````

## 4 - Counting with a For Loop

Once we know how many rounds to run, we use a `for` loop to repeat that many times.

```python
for i in range(1, num_flips + 1):
    # runs once per flip
    # i = 1, then 2, then 3... up to num_flips
```

### Why `range(1, num_flips + 1)`?
- `range(start, stop)` - goes from `start` up to (but NOT including) `stop`
- Starting at `1` lets us use `i` as the round number in our output
- We add `+1` to the stop so we actually get `num_flips` iterations

### Tracking a running total
Inside the loop, we can keep a **counter variable** to track how many times something happens:

```python
heads_count = 0           # start at 0 - BEFORE the loop

for i in range(1, num_flips + 1):
    result = flip_coin()
    if result == "Heads":
        heads_count += 1  # add 1 every time we see Heads

return heads_count        # send the total back - AFTER the loop
```

````python
import random

def flip_coin():
    if random.randint(1, 2) == 1:
        return "Heads"
    return "Tails"

# --- The game loop ---
def play_game(num_flips):
    heads_count = 0                        # counter starts at 0

    for i in range(1, num_flips + 1):      # repeat num_flips times
        result = flip_coin()               # call the function each round
        print(f"  Flip {i}: {result}")
        if result == "Heads":
            heads_count += 1               # count heads

    return heads_count                     # return the total to main()

def main():
    heads = play_game(5)                   # run 5 flips
    print(f"\nTotal Heads: {heads}")

main()
````

## 5 - The Full Program

Now let's put all four pieces together.

### Program structure for Assignment 3

```
main()
  │
  ├─── get_num_flips()      ← validates input, returns an integer
  │
  ├─── play_game(n)         ← runs the loop, returns a score/count
  │         │
  │         └── flip_coin() ← generates one random result, returns it
  │
  └─── prints final results
```

**Key rules to remember:**
- `flip_coin()` → returns a value, no printing
- `get_num_flips()` → returns a value, no printing (except error messages)
- `play_game()` → returns a value (the score), printing each round is OK here
- `main()` → calls everything, prints the final summary

---
The cell below is the **complete working solution** for the Coin Flip Tracker.  
Notice how each function does exactly one job.

````python
'''
Name:        Ms. Bean (Demo)
Date:        March 2026
Assignment:  Coin Flip Tracker - Teaching Demo
Description: Flips a coin a user-chosen number of times and
             reports the results. Demonstrates random numbers,
             functions, loops, and input validation.
'''

import random


# --- Flips one coin. Returns "Heads" or "Tails". ---
def flip_coin():
    number = random.randint(1, 2)   # 1 = Heads, 2 = Tails
    if number == 1:
        return "Heads"
    else:
        return "Tails"


# --- Asks how many flips (1–10). Validates input. Returns integer. ---
def get_num_flips():
    while True:
        user_input = input("How many flips would you like? (1–10): ")
        if user_input.isdigit():
            num = int(user_input)
            if 1 <= num <= 10:
                return num
        print("  Please enter a whole number between 1 and 10.")


# --- Runs the flip loop. Returns the number of Heads. ---
def play_game(num_flips):
    heads_count = 0

    for i in range(1, num_flips + 1):
        result = flip_coin()
        print(f"  Flip {i}: {result}")
        if result == "Heads":
            heads_count += 1

    return heads_count


# --- Controls the program. All final output lives here. ---
def main():
    print("================================")
    print("      COIN FLIP TRACKER         ")
    print("================================\n")

    num_flips = get_num_flips()
    print()

    heads = play_game(num_flips)
    tails = num_flips - heads

    print(f"\n--- Results ---")
    print(f"Heads : {heads}")
    print(f"Tails : {tails}")

    if heads > tails:
        print("\nHeads takes the win!")
    elif tails > heads:
        print("\nTails takes the win!")
    else:
        print("\nIt's a perfect tie!")


main()
````

## 6 - Bonus: Scoring with `if / elif / else`

Your assignment needs a scoring system based on what was rolled.  
Here is the general pattern using coin flips as an example:

```python
def calculate_score(result_a, result_b):
    if result_a == "Heads" and result_b == "Heads":   # both match - big points
        return 50
    elif result_a == "Tails" and result_b == "Tails": # both tails - smaller points
        return 20
    else:                                              # split result
        return 10
```

The order of your `if / elif` conditions matters - Python checks them **top to bottom** and stops at the first match.

For your assignment scoring, think about:
- What's the rarest / best outcome? → check it **first**
- What conditions overlap? → handle the most specific ones before the general ones

````python
import random

def flip_coin():
    if random.randint(1, 2) == 1:
        return "Heads"
    return "Tails"

# --- A scoring function that checks both flips ---
def calculate_score(flip1, flip2):
    if flip1 == "Heads" and flip2 == "Heads":   # best outcome - check first!
        return 50
    elif flip1 == "Tails" and flip2 == "Tails":
        return 20
    else:
        return 10

def main():
    total_score = 0

    for i in range(1, 4):                           # 3 rounds
        f1 = flip_coin()
        f2 = flip_coin()
        points = calculate_score(f1, f2)            # pass both results in
        total_score += points
        print(f"Round {i}: {f1} + {f2} → {points} points")

    print(f"\nTotal Score: {total_score}")

main()
````
