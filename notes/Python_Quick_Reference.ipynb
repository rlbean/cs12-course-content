## Output Formatting with f-strings

### Basic f-string

````python
name = "Jordan"
age = 16
print(f"I'm {name} and I'm {age}")
# Output: I'm Jordan and I'm 16
````

---
### Format Numbers - 2 Decimals

````python
price = 9.5
print(f"Price: ${price:.2f}")
# Output: Price: $9.50
````

---
### Format Numbers - Commas

````python
big = 1234567
print(f"Total: {big:,}")
# Output: Total: 1,234,567
````

---
### Math Inside f-strings

````python
length = 10
width = 5
print(f"Area: {length * width} square feet")
# Output: Area: 50 square feet
````

## Common Patterns
### Get, Calculate, Display

````python
# 1. Get input
radius = float(input("Radius? "))

# 2. Calculate
area = 3.14159 * radius ** 2

# 3. Display
print(f"Area: {area:.2f} square units")
````

---
### Multiple Calculations

````python
length = float(input("Length? "))
width = float(input("Width? "))

perimeter = 2 * (length + width)
area = length * width

print(f"Perimeter: {perimeter:.2f}")
print(f"Area: {area:.2f}")
````

### Type Conversion Reference

````python
# String to integer
int("42")         # → 42
int("3.14")       # ✗ ERROR!

# String to float
float("3.14")     # → 3.14
float("42")       # → 42.0

# Number to string
str(42)           # → "42"
str(3.14)         # → "3.14"
````

## Common Mistakes & Fixes
### Forgot to convert input -- WRONG

````python
age = input("Age? ")
print(age + 1)  # ERROR!
````

age = input("Age? ")
print(age + 1)  # ERROR!age = input("Age? ")
print(age + 1)  # ERROR!

### Fixed

````python
age = int(input("Age? "))
print(age + 1)  # Works!
````

### Used int() for decimals -- WRONG

````python
name = "Sam"
print("Hi {name}")  # Prints: Hi {name}
````

### Fixed

````python
print(f"Hi {name}")  # Prints: Hi Sam
````

### Math on strings - WRONG

````python
num1 = "5"
num2 = "3"
print(num1 + num2)  # Prints: 53
````

### Fixed

````python
num1 = int("5")
num2 = int("3")
print(num1 + num2)  # Prints: 8
````
